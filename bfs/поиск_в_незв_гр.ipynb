```python
# Определим константы
INF = int(1e9)

# Функция для инициализации матрицы расстояний
def initialize_dist(vertex_count):
    return [[INF] * vertex_count for _ in range(vertex_count)]

# Основная функция
def floyd_warshall(vertex_count, edge_count, edges):
    # Инициализация матрицы расстояний
    dist = initialize_dist(vertex_count)
    
    # Устанавливаем расстояние до самих себя равным 0
    for v in range(vertex_count):
        dist[v][v] = 0
    
    # Заполнение матрицы distances в соответствии с ребрами
    for edge in edges:
        a, b, weight = edge
        dist[a][b] = min(dist[a][b], weight)

    # Алгоритм Флойда-Уоршелла
    for v in range(vertex_count):
        for a in range(vertex_count):
            for b in range(vertex_count):
                if dist[a][v] != INF and dist[v][b] != INF and dist[a][b] > dist[a][v] + dist[v][b]:
                    dist[a][b] = dist[a][v] + dist[v][b]
    
    return dist

# Пример использования
vertex_count = 5  # количество вершин
edge_count = 6    # количество рёбер
edges = [
    (0, 1, 10),
    (0, 2, 3),
    (1, 2, 1),
    (1, 3, 2),
    (2, 1, 4),
    (2, 3, 8),
    (3, 4, 7),
]

distance_matrix = floyd_warshall(vertex_count, edge_count, edges)

# Выводим матрицу расстояний
for row in distance_matrix:
    print(row)
```
