```python
import matplotlib.pyplot as plt
import math


axiom = "F"
rules = {
    "F": "-F+F+[+F-F-]-[-F+F+F]"
}
iterations = 4
angle = -math.pi / 8


# L-
def l_system(axiom, rules, n):
    for _ in range(n):
        next_axiom = ""
        for char in axiom:
            next_axiom += rules.get(char, char)
        axiom = next_axiom
    return axiom

#   
def gosper_curve(instruction, angle):
    x, y = 0, 0
    stack = []
    coords = [(x, y)]
    current_angle = 0

    for cmd in instruction:
        if cmd == 'F':
            x += math.cos(current_angle)
            y += math.sin(current_angle)
            coords.append((x, y))
        elif cmd == '+':
            current_angle += angle
        elif cmd == '-':
            current_angle -= angle
        elif cmd == '[':
            stack.append((x, y, current_angle))
        elif cmd == ']':
            x, y, current_angle = stack.pop()
            coords.append((x, y))

    return coords

# 
instructions = l_system(axiom, rules, n)
coords = gosper_curve(instructions, angle)

x_coords, y_coords = zip(*coords)

plt.figure(figsize=(8, 8))
plt.plot(x_coords, y_coords, color='black')
plt.axis('equal')
plt.axis('off')
plt.title('')
plt.show()
```

```python
import turtle
import math

#   L-
axiom = "F"
rules = {
    "F": "-F+F+[+F-F-]-[-F+F+F]"
}
iterations = 6
angle = -math.pi / 8

#     L-
def generate_l_system(axiom, rules, iterations):
    current_string = axiom
    for _ in range(iterations):
        next_string = ""
        for char in current_string:
            next_string += rules.get(char, char)  #    
        current_string = next_string
    return current_string

#    
def draw_l_system(t, instructions, angle):
    stack = []
    for command in instructions:
        if command == "F":
            t.forward(5)  #  
        elif command == "+":
            t.right(math.degrees(angle))
        elif command == "-":
            t.left(math.degrees(angle))
        elif command == "[":
            stack.append((t.position(), t.heading()))
        elif command == "]":
            position, heading = stack.pop()
            t.penup()
            t.setposition(position)
            t.setheading(heading)
            t.pendown()

#  
turtle.speed(0)  #   ()
turtle.delay(0)  #    

#  L-
l_system_string = generate_l_system(axiom, rules, iterations)

#  
draw_l_system(turtle, l_system_string, angle)

#  
turtle.done()
```
