```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры
seed = complex(0.56667, 0)
power1 = 2
power2 = 0
induct = complex(-0.5, 0)
bailout = 1.0e20

# Размеры изображения
width, height = 800, 800
x_min, x_max = -2, 2
y_min, y_max = -2, 2

# Создаем массив для хранения значений
image = np.zeros((height, width))

for i in range(height):
    for j in range(width):
        z = complex(0, 0)  # Начальное значение z
        x = x_min + (x_max - x_min) * j / (width - 1)
        y = y_min + (y_max - y_min) * i / (height - 1)
        c = complex(x, y)
        
        # Итерации
        iteration = 0
        while abs(z) < bailout and iteration < 1000:  # Ограничиваем количество итераций
            newz = z ** power1 + c * seed + induct * z
            z = newz
            iteration += 1
        
        # Сохраняем количество итераций для цветовой карты
        image[i, j] = iteration

# Визуализация
plt.imshow(image, extent=(x_min, x_max, y_min, y_max), cmap='hot')
plt.colorbar()
plt.title('Fractal Visualization')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```
