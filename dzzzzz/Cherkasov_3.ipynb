```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры модели
r = 3  # параметр роста
x0 = 0.02  # начальная численность популяции
iterations = 70

# Инициализация списков для значений
x_values = [x0]
x_next_values = []

# Вычисление значений численности популяции
for i in range(iterations):
    x_next = x_values[i] + r * x_values[i] * (1 - x_values[i])
    x_values.append(x_next)
    x_next_values.append(x_next)

# Построение графика зависимости численности популяции от итераций
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(iterations + 1), x_values, 'o-', label="Популяция")
plt.xlabel("Итерации")
plt.ylabel("Численность популяции")
plt.title("График зависимости численности популяции от итераций")
plt.grid(True)
plt.legend()

# Построение фазового портрета
plt.subplot(1, 2, 2)
plt.plot(x_values[:-1], x_next_values, 'o-', label="Фазовый портрет")
plt.xlabel("x_m")
plt.ylabel("x_(m+1)")
plt.title("Фазовый портрет")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
