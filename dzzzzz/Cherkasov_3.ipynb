```python
import numpy as np
import matplotlib.pyplot as plt

#  
r = 3  #  
x0 = 0.02  #   
iterations = 70

#    
x_values = [x0]
x_next_values = []

#    
for i in range(iterations):
    x_next = x_values[i] + r * x_values[i] * (1 - x_values[i])
    x_values.append(x_next)
    x_next_values.append(x_next)

#       
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(iterations + 1), x_values, 'o-', label="")
plt.xlabel("")
plt.ylabel(" ")
plt.title("     ")
plt.grid(True)
plt.legend()

#   
plt.subplot(1, 2, 2)
plt.plot(x_values[:-1], x_next_values, 'o-', label=" ")
plt.xlabel("x_m")
plt.ylabel("x_(m+1)")
plt.title(" ")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
